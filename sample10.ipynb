```python
import re  # Add this at the top of your script for regex processing

def normalize_text(text):
    """Removes special symbols and extra spaces while preserving important content."""
    text = re.sub(r"[^\w\s,]", "", text)  # Remove non-word characters except spaces and commas
    text = re.sub(r"\s+", " ", text).strip()  # Normalize extra spaces
    return text
```

```python
import requests
import json
import pandas as pd
import time
import os

# === CONFIGURATION ===
df_sample = df.head(10).copy() 
INPUT_FILE = df_sample  # Input file with comments
OUTPUT_FILE = "processed_comments.csv"  # Output file to store results
BATCH_SIZE = 10                # Save progress every 100 comments
API_URL = "http://localhost:1234/v1/completions"  # LM Studio API endpoint
MODEL_NAME = "mistral-7b-instruct@q3_k_l"  # Model in LM Studio
MAX_RETRIES = 5                 # Number of retries for API errors
PROMPT_DELAY = 0.5           # Time (seconds) between each API request

# === LOAD DATA & RESUME IF NECESSARY ===
try: 
    processed_df = pd.read_csv(OUTPUT_FILE)  
    last_index = len(processed_df) - 1  
    print(f"Resuming from index {last_index + 1}...")
except FileNotFoundError:
    df_sample["label"] = ""  
    last_index = -1  

# === FUNCTION TO HANDLE API REQUESTS WITH RETRIES ===
def correct_spelling_deepseek(text):
    """Calls LM Studio API to correct spelling and label the comment."""
    prompt = f"""
You are a data labeler and a spell checker with the main task of interpreting social media comments on trending YouTube Shorts videos from Indonesia and assigning a label to each of them. 
Your task is to spellcheck and interpret the meaning of the social media comments. 
Label the comment as either "Online Gambling Promotion" if it promotes online gambling, or "No Online Gambling Promotion" otherwise.

Correct the spelling in this Indonesian text if it is not understandable. Preserve gambling terms like 'judi', 'slot', 'bonus', 
'gachor', 'maxwin', or any other keywords that might indicate online gambling promotions. 

**Output strictly in this format:**  
`<corrected_comment>, <label>` 
Do not add punctuations to the comments only add them to differentiate comment and label

Don't give any explanationsâ€”just output the corrected or uncorrected comment along with its label. Make sure the output follows this format:
Corrected comment without parentheses: "<corrected_comment>, <label>"

Example of Expected Format:
1. Input : sy mau daptar judi onlen -> "saya mau daftar judi online, Online Gambling Promotion"
2. Input : bersandar teras rumah sembari ngopi bermain permainan nyelot jptogel77 nikmat sekaleh â†’ 
    "bersandar teras rumah sembari ngopi bermain permainan slot jptogel77 nikmat sekali, Online Gambling Promotion"
3. Input : lagi nunggu ojol di pinggir jalan â†’  "lagi nunggu ojol di pinggir jalan, No Online Gambling Promotion"

Output format enforcement: **Do NOT include explanations or extra words. Only return the formatted result.**  

Text to process: "{text}"
"""

    payload = {
        "model": MODEL_NAME,
        "prompt": prompt,
        "max_tokens": 80,
        "temperature": 0.3
    }

    headers = {"Content-Type": "application/json"}

    for attempt in range(MAX_RETRIES):
        try:
            response = requests.post(API_URL, headers=headers, json=payload)
            response_json = response.json()
            
            # Extract the generated response
            corrected_text = response_json["choices"][0]["text"].strip()
            return corrected_text if corrected_text else text  

        except Exception as e:
            print(f"Error on attempt {attempt+1}: {e}")
            time.sleep(2 ** attempt)  # Exponential backoff (2s, 4s, 8s...)

    return text  

# === PROCESS COMMENTS ===
for i in range(last_index + 1, len(df_sample)):
    comment = df_sample.loc[i, "comment"]
    processed_text = correct_spelling_deepseek(comment).replace("\\", "").replace('"',"")
    # print(correct_spelling_deepseek(comment))
    print(processed_text)
    try:
        processed_text = normalize_text(processed_text).strip('"')  # Apply normalization
        print(f"processed_text : {processed_text}")
        corrected_comment, label = processed_text.rsplit(",", 1)
        corrected_comment = corrected_comment.strip()
        label = label.strip()
    except ValueError:
        corrected_comment, label = processed_text.strip(), "ERROR"

    df_sample.loc[i, "comment"] = corrected_comment  
    df_sample.loc[i, "label"] = label

    # Save progress every BATCH_SIZE comments
    if (i + 1) % BATCH_SIZE == 0 or i == len(df_sample) - 1:
        df_sample.iloc[:i+1].to_csv(OUTPUT_FILE, index=False)
        print(f"âœ… Saved progress at {i+1} comments")

    time.sleep(PROMPT_DELAY)  # === Added Delay Between Requests ===
    print()
print("ðŸŽ‰ Processing complete! Data saved.")
```
